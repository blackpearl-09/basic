## torch 가 설치되지 않을 때 

1. Anaconda Prompt를 관리자 권한으로 실행
2. 파이썬 버전 확인을 위해 python --version 입력

3. pytorch를 설치할 새로운 가상환경 생성 
* 입력할 내용: conda create -n 가상환경이름
 -**-n 뒤에는 정할 명칭(이름)을 입력**-

-**결과**- y 입력

5. 생성한 가상환경에 들어가기 
 * conda activate Pytorch

6. 그 전에 내가 설치한 가상환경 목록을 확인하고 싶다면, 
 * conda env list 입력

7. 위 Anaconda Prompt에서 (base) -> (Pytorch)로 변경된 걸 볼 수 있음
8. 이렇게 만든 가상환경에 pytorch를 설치하기 위해선,pytorch 사이트 (https://pytorch.org/get-started/locally/) 접속
